```python
from datetime import datetime
import csv
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Sitka Weather Data
path = Path('./sitka_weather_2021_full.csv')
lines = path.read_text().splitlines()

reader = csv.reader(lines)
header_row = next(reader)
```

```python
# Sitka Weather Data
path2 = Path('./death_valley_2021_full.csv')
lines2 = path2.read_text().splitlines()

reader2 = csv.reader(lines2)
header_row2 = next(reader2)
```

```python
# Checks indexes for Sitka Weather Data
for index, column_header in enumerate(header_row):
    print(index,column_header)
```

```python
# Checks indexes for Sitka Weather Data
for index, column_header in enumerate(header_row2):
    print(index,column_header)
```

```python
dates, highs = [], []
dates2, highs2 = [], []

for row in reader:
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    try:
        high = int(row[7])
    except ValueError:
        print(f'No Data for {current_date}')
    else:
        dates.append(current_date)
        highs.append(high)
for row in reader2:
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    try:
        high = int(row[7])
    except ValueError:
        print(f'No Data for {current_date}')
    else:
        dates2.append(current_date)
        highs2.append(high)
```

```python
fig, ax = plt.subplots()
ax.plot(dates,highs,color='red')
ax.plot(dates2,highs2,color='blue')

ax.set_title('Temperature Data in Sitka and Death Valley 2021')
ax.set_ylabel("Temperature in (F)",fontsize=16)
ax.set_xlabel("Date")
ax.legend(['Sitka Temperature', 'Death Valley'])
fig.autofmt_xdate()

plt.show()
```
