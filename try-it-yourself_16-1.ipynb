```python
from datetime import datetime
import csv
from pathlib import Path
import matplotlib.pyplot as plt


```

```python
path = Path('./sitka_weather_2021_full.csv')
lines = path.read_text().splitlines()

reader = csv.reader(lines)
header_row = next(reader)
```

```python
for index, column_header in enumerate(header_row):
    print(index,column_header)
```

```python
fig, ax = plt.subplots()
dates,precipitations = [],[]

for row in reader:
    print(row[5])
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    try:
        precipitation = float(row[5])
    except ValueError:
        print(f'No Data for {current_date}')
    else:
        dates.append(current_date)
        precipitations.append(precipitation)
print(precipitations)
ax.plot(dates,precipitations)
ax.set_title('Precipitation Data in Sitka 2021')
ax.set_ylabel("Chance of Rain (PP)",fontsize=16)
ax.set_xlabel("Date")
fig.autofmt_xdate()

plt.show()
```
